plot() 最少必要知识：
> 1. 输入两个数组，第一个数组代表x，第二个数组代表y（第一个数组可以忽略不写，此时x的取值为0,1,...,len(y)-1;返回一个各个(x, y)连线而成的图像。
> 2. 可以添加多个参数format(样式) -- \**arg(位置参数)
> 3. 可以添加Line2D中的property(特性) -- \**karg(关键字参数) P.S: 如果format和property冲突，则以property为准。
> 4. 想画多条线时可以： ```plot(\[x1\], y1, \[fmt1\], \[x2\], y2, \[fmt2\], ...)``` 或者多次使用plot()函数。前者可以定制每条线的format,但是会共享Line2D的property。
> 5. 

````python
from matplotlib.pyplot import *

x = [1, 2, 3, 4]
y = [[1, 2], [2, 3], [3, 4, 6, 6], [4, 5]]

plot(y[0], y[1])
plot(x, y[2])
````
